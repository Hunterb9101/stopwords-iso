```python
import stopwordsiso as stopwords

# We use "stopwordsiso" as a package name instead of "stopwords-iso",
# as package name in Python should not has an hyphen.
```

```python
stopwords.has_lang("th") # check if there is a stopwords list for the language
```

```python
stopwords.list_lang() # list all the supported languages
```

```python
stopwords.stopwords("en") # get a list of stopwords in a specific language
```

```python
stopwords.stopwords(["ja", "th", "en"]) # get a list of stopwords in multiple languages
```

```python
stopwords.stopwords("xxx") # if the language is not supported, return an empty list
```
